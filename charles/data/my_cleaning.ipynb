```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

# Droping irrelevant columns
PID, Prop_addr dont play into price, mean lot frontage was a clumn we made to impute some miussing values during data cleaning, as well as lotBucket, and garage Year Built has so many None values that we dont want to impute since it'd be straight up lying

```python
data.drop(['PID','Prop_Addr','mean_LotFrontage','lot_bucket','GarageYrBlt'],axis = 1, inplace = True)
outliers = [115, 132, 141, 335, 490, 538, 2047, 2239]
data.drop(outliers, inplace = True)
```

# Making a column for the number of bathrooms in a house

```python
# Combining all the Bathroom Columnsb
data['CombinedBathroom'] = data['BsmtFullBath'] + data['FullBath']+  .5*data['BsmtHalfBath'] + .5*data['HalfBath']
data.drop(['BsmtFullBath','FullBath','BsmtHalfBath','HalfBath'], inplace = True, axis = 1)
```

# Adding a feature Tony said was nice, Tonys feature

```python
data['TonyFeature'] = data['OverallCond']*data['OverallQual']
```

# Log GrLivArea

```python
data['logGrLivArea'] = np.log(data.GrLivArea)
```
