```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
pd.set_option('display.max_columns',None)
from joblib import dump, load
```

```python
data = pd.read_csv('./../../../communal/Ames_Housing_Price_Data_cleaned_2.csv')
data = data.drop(['PID','Prop_Addr','mean_LotFrontage'],axis = 1)
outliers = [115, 132, 141, 335, 490, 538, 2047, 2239]
data.drop(outliers, inplace = True)
predict = data['SalePrice']
log_predict = np.log(predict)
data.drop(['SalePrice','lot_bucket','lat','long','GarageYrBlt'],axis = 1, inplace = True)
dummy = pd.get_dummies(data,drop_first = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dummy, log_predict)
```

# Using Grid Search to find the best RandomForest on All Data

```python
params = {
    'max_depth' : range(4,20),
    'n_estimators': range(100,1000,100),
    'min_samples_split': range(10,60,10),
    
}

gridSearchForest = GridSearchCV(RandomForestRegressor(oob_score=True, max_features = 'sqrt'), param_grid=params, n_jobs = -1, verbose=3)

gridSearchForest.fit(dummy,log_predict)
```

```python
best_forest = gridSearchForest.best_estimator_
dump(best_forest, 'best_random_forest_on_all_data.joblib')
```

```python
gridSearchForest.fit(X_train, y_train)
```

```python
best_forest = gridSearchForest.best_estimator_
print(best_forest.score(X_train, y_train))
print(best_forest.score(X_test, y_test))
residuals = best_forest.predict(X_test) - y_test
dump(best_forest, 'best_random_forest_on _split_data.joblib')
```
